FOR TRAINING DATASET

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load train dataset
train_data = pd.read_csv('/content/train_.csv')

# Convert dates to datetime format
train_data['order_delivered_date'] = pd.to_datetime(train_data['order_delivered_date'])
train_data['order_shipped_date'] = pd.to_datetime(train_data['order_shipped_date'])

# Calculate delivery time
train_data['delivery_time'] = (train_data['order_delivered_date'] - train_data['order_shipped_date']).dt.days

# 1. Delivery SLA vs Actual Delivery Time
plt.figure(figsize=(10, 6))
sns.scatterplot(x=train_data['order_delivery_sla'], y=train_data['delivery_time'], alpha=0.5)
plt.title('Order Delivery SLA vs Actual Delivery Time')
plt.xlabel('Order Delivery SLA (Days)')
plt.ylabel('Actual Delivery Time (Days)')
plt.show()

# 2. Distribution of Account Modes
plt.figure(figsize=(8, 6))
sns.countplot(x='account_mode', data=train_data, palette='viridis')
plt.title('Distribution of Account Modes')
plt.xlabel('Account Mode')
plt.ylabel('Count')
plt.show()

# 3. Distribution of Delivery Time
plt.figure(figsize=(8, 6))
sns.histplot(train_data['delivery_time'], kde=True, bins=30, color='blue')
plt.title('Distribution of Delivery Time')
plt.xlabel('Days')
plt.ylabel('Frequency')
plt.show()
```

FOR PINCODES DATASET

```python
# Load pincodes dataset
pincodes_data = pd.read_csv('/content/pincodes_cleaned.csv')

# 1. Top 10 States with the Most Pincodes
top_states = pincodes_data['StateName'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_states.index, y=top_states.values, palette='coolwarm')
plt.title('Top 10 States with Most Pincodes')
plt.xlabel('State')
plt.ylabel('Number of Pincodes')
plt.xticks(rotation=45)
plt.show()

# 2. Latitude vs Longitude Distribution
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Longitude', y='Latitude', data=pincodes_data, alpha=0.3, color='purple')
plt.title('Latitude vs Longitude of Pincodes')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

FOR TESTING DATASET

```python
# Load test dataset
test_data = pd.read_csv('/content/test_.csv')

# 1. Distribution of Account Modes
plt.figure(figsize=(8, 6))
sns.countplot(x='account_mode', data=test_data, palette='viridis')
plt.title('Distribution of Account Modes (Test Data)')
plt.xlabel('Account Mode')
plt.ylabel('Count')
plt.show()

# 2. Courier Partner vs Drop Pincode
top_couriers = test_data['courier_partner_id'].value_counts().head(10).index
filtered_data = test_data[test_data['courier_partner_id'].isin(top_couriers)]

plt.figure(figsize=(12, 6))
sns.boxplot(x='courier_partner_id', y='drop_pin_code', data=filtered_data, palette='coolwarm')
plt.title('Courier Partner vs Drop Pincode')
plt.xlabel('Courier Partner ID')
plt.ylabel('Drop Pincode')
plt.show()
```
